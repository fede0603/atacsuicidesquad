```python
import pandas as pd
import os
import numpy
```

```python
# Define the folder path where your CSV files are located
folder_path = '../feeddata/'
```

```python


# Get a list of all files in the folder
files = os.listdir(folder_path)

# Filter files that start with 'tu' and end with '.csv'
tu_csv_files = [file for file in files if file.startswith('tu') and file.endswith('.csv')]

# Initialize an empty DataFrame to store the concatenated data
df_tu_total = pd.DataFrame()

dtypes = {'id': str,
 'isDeleted': str,
 'tripUpdate.trip.tripId': str,
 'tripUpdate.trip.startTime': str,
 'tripUpdate.trip.startDate': str,
 'tripUpdate.trip.routeId': str,
 'tripUpdate.trip.directionId': str,
 'tripUpdate.stopTimeUpdate': str,
 'tripUpdate.vehicle.id': str,
 'tripUpdate.vehicle.label': str,
 'tripUpdate.timestamp': int,
 'tripUpdate.delay': float,
 'max_time': int}

# Loop through each CSV file and concatenate them
for file in tu_csv_files:
    file_path = os.path.join(folder_path, file,  )
    data = pd.read_csv(file_path, sep = ";", dtype = dtypes)  # Read the CSV file
    df_tu_total = pd.concat([df_tu_total, data], ignore_index=True)  # Concatenate the data
```

```python
# Convert Unix timestamp to datetime
df_tu_total['tripUpdate.timestamp'] = pd.to_datetime(df_tu_total['tripUpdate.timestamp'], unit='s')
df_tu_total['max_time'] = pd.to_datetime(df_tu_total['max_time'], unit='s')

df_tu_total.to_csv('./output_data/tu_feed.csv', index=False)
```
