```python
routes = pd.read_csv(r"./rome_static_gtfs/routes.txt", sep = ",")
atac_routes = routes.loc[routes["agency_id"] == "OP1"][["route_id","route_short_name","route_long_name","route_type"]]
atac_routes = atac_routes["route_id"].drop_duplicates()
atac_routes.head()
```

```python
len(routes), len(atac_routes)
```

```python
dtypes = {
    'route_id': str,
    'service_id': str,
    'trip_id': str,
    'trip_headsign': str,
    'trip_short_name': str,
    'direction_id': str,
    'block_id': str,
    'shape_id': str,
    'wheelchair_accessible': str,
    'exceptional': str
}
trips = pd.read_csv(r"./rome_static_gtfs/trips.txt", sep = ",")[["route_id","service_id","trip_id","trip_headsign","block_id"]]
atac_trips = trips.merge(atac_routes, on = "route_id", how = "inner")
atac_trips
```

```python
len(trips),len(atac_trips)
```

```python
dtypes = {"service_id": str}
calendar = pd.read_csv(r"./rome_static_gtfs/calendar.txt", sep = ",", dtype = dtypes)
# Define the date range
start_date_range = "2023-09-06"
end_date_range = "2023-09-06"

# Convert int64 to datetime
calendar['start_date'] = pd.to_datetime(calendar['start_date'], format='%Y%m%d')
calendar['end_date'] = pd.to_datetime(calendar['end_date'], format='%Y%m%d')

# Convert date range to datetime objects
start_date_range = pd.to_datetime(start_date_range)
end_date_range = pd.to_datetime(end_date_range)

# Filter the DataFrame
filtered_calendar = calendar.loc[
    (calendar["start_date"] <= start_date_range) &
    (calendar["end_date"] >= end_date_range) &
(calendar["wednesday"] == 1)
][["service_id"]]
```

```python
active_atac_trips = atac_trips.merge(filtered_calendar, on = "service_id", how = "inner")
```
